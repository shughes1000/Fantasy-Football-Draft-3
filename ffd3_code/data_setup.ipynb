# Data Setup

### ADP Data

```python
import pandas as pd

df_ADP = pd.read_html("https://www.fantasypros.com/nfl/adp/ppr-overall.php?year=2012")[0]
df_ADP["year"] = int(2012)
df_ADP.head()
```

```python
for i in range(13, 22):
    df_ADP = pd.concat([df_ADP, pd.read_html(
        "https://www.fantasypros.com/nfl/adp/ppr-overall.php?year=20" + str(i))[0]])
    
    df_ADP.loc[df_ADP["year"].isnull(), "year"] = int("20" + str(i))
    
df_ADP.reset_index(drop = True, inplace = True)
df_ADP['year'] = df_ADP['year'].astype(int)

df_ADP.sample(10)
```

```python
df_ADP.rename({
               "Rank": "rank_adp", 
               "Player Team (Bye)": "player", 
               "AVG": "avg_adp"
              }, axis = 1, inplace = True)
df_ADP.columns = df_ADP.columns.str.lower()
df_ADP.head()
```

```python
df_ADP['pos'] = df_ADP['pos'].str.replace('\d+', '')
df_ADP.head()
```

```python
df_ADP.loc[df_ADP["pos"] == "DST", "player"] = [i[0] for i in 
                                                df_ADP.loc[df_ADP["pos"] == "DST", 
                                                           "player"].str.split(" DST")]

temp = df_ADP.loc[df_ADP["pos"] != "DST", "player"].str.split(n = 2)
df_ADP.loc[df_ADP["pos"] != "DST", "player"] = temp.str[0] + " " + temp.str[1]
df_ADP.head()
```

### Points Data

```python
df_points = pd.read_html(
    "https://www.fantasypros.com/nfl/reports/leaders/ppr.php?year=2012&start=1&end=16")[0]
df_points["year"] = 2012
df_points.head()
```

```python
for i in range(13, 22):
    df_points = pd.concat([df_points, 
                           pd.read_html(
                               ("https://www.fantasypros.com/nfl/reports/leaders/ppr.php?year=20" 
                                + str(i) + "&start=1&end=16")
                           )[0]])
    df_points.loc[df_points["year"].isnull(), "year"] = int("20" + str(i))
    
df_points.reset_index(drop = True, inplace = True)
df_points['year'] = df_points['year'].astype(int)

df_points.sample(10)
```

```python
df_points.rename({
                  "Rank": "rank_scoring", 
                  "Position": "pos", 
                  "Avg": "avg_scoring"
                 }, axis = 1, inplace = True)
df_points.columns = df_points.columns.str.lower()
df_points.head()
```

```python
temp = df_points.loc[df_points["pos"] != "DST", "player"].str.split(n = 2)
df_points.loc[df_points["pos"] != "DST", "player"] = temp.str[0] + " " + temp.str[1]
df_points.head()
```

### Merge and Save

```python
df = pd.merge(df_ADP, df_points,  how='left', on = ["player", "pos", "year"])
df.head()
```

Because drafts typically only last until pick 225 at the latest, we should remove all occurences of draft ranking above 250. Players who were injured should also be removed because they may have misleading averages. This is due to the small sample size of games played.

```python
df = df.loc[(df["rank_adp"] <= 250) & (df["games"] > 8), ]
df.sample(5)
```
